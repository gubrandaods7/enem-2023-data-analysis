## Quantidade de participantes por sexo (Masculino vs Feminino)

```python
query = """
SELECT TP_SEXO, COUNT(NU_INSCRICAO) AS Quantidade, ROUND(COUNT(NU_INSCRICAO) * 100.0 / SUM(COUNT(NU_INSCRICAO)) OVER(), 2) AS Percentual
FROM enem
GROUP BY TP_SEXO
ORDER BY Quantidade DESC
"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## Média das notas por tipo de prova e redação

```python
query = """
SELECT 
    ROUND(AVG(NU_NOTA_CN),2) AS Media_CN,
    ROUND(AVG(NU_NOTA_CH),2) AS Media_CH,
    ROUND(AVG(NU_NOTA_LC),2) AS Media_LC,
    ROUND(AVG(NU_NOTA_MT),2) AS Media_MT,
    ROUND(AVG(NU_NOTA_REDACAO),2) AS Media_Redacao,
    ROUND(AVG((NU_NOTA_CN + NU_NOTA_CH + NU_NOTA_LC + NU_NOTA_MT + NU_NOTA_REDACAO)/5.0),2) AS Media_Geral
FROM enem
WHERE NU_NOTA_CN IS NOT NULL 
  AND NU_NOTA_CH IS NOT NULL
  AND NU_NOTA_LC IS NOT NULL
  AND NU_NOTA_MT IS NOT NULL
  AND NU_NOTA_REDACAO IS NOT NULL;

"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## Média de nota geral por estado

```python
query = """
SELECT 
    SG_UF_PROVA,
    ROUND(AVG((NU_NOTA_CN + NU_NOTA_CH + NU_NOTA_LC + NU_NOTA_MT + NU_NOTA_REDACAO)/5.0), 2) AS Nota_Media
FROM enem
WHERE NU_NOTA_CN IS NOT NULL 
  AND NU_NOTA_CH IS NOT NULL
  AND NU_NOTA_LC IS NOT NULL
  AND NU_NOTA_MT IS NOT NULL
  AND NU_NOTA_REDACAO IS NOT NULL
GROUP BY SG_UF_PROVA
ORDER BY Nota_Media DESC;

"""

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```

## Como a renda se distribui dentro de cada raça

# Q006 renda mensal
# TP_COR_RACA 

query = """
SELECT 
    TP_COR_RACA,
    Q006,
    COUNT(*) AS Quantidade,
    ROUND(
    -- Qual a porcentagem de pessoas de cada faixa de renda, dentro de cada raça
        COUNT(*) * 100.0 / SUM(COUNT(*)) OVER (PARTITION BY TP_COR_RACA), 
        2
    ) AS Percentual
FROM enem
WHERE Q006 IS NOT NULL
  AND TP_COR_RACA IS NOT NULL
GROUP BY TP_COR_RACA, Q006
ORDER BY TP_COR_RACA, Percentual DESC;

"""
pd.set_option("display.max_rows", None)

df_resultado = pd.read_sql_query(query, conn)
df_resultado

## Média de nota por tipo de ensino (Público, Privado)

```python
# Q006 renda mensal
# TP_COR_RACA 

query = """
SELECT 
    TP_ESCOLA,
    ROUND(AVG((NU_NOTA_CN + NU_NOTA_CH + NU_NOTA_LC + NU_NOTA_MT + NU_NOTA_REDACAO)/5.0), 2) AS Media_Geral,
    ROUND(AVG(NU_NOTA_CN),2) AS Media_CN,
    ROUND(AVG(NU_NOTA_CH),2) AS Media_CH,
    ROUND(AVG(NU_NOTA_LC),2) AS Media_LC,
    ROUND(AVG(NU_NOTA_MT),2) AS Media_MT,
    ROUND(AVG(NU_NOTA_REDACAO),2) AS Media_Redacao,
    COUNT(*) AS Quantidade
FROM enem
WHERE NU_NOTA_CN IS NOT NULL 
  AND NU_NOTA_CH IS NOT NULL
  AND NU_NOTA_LC IS NOT NULL
  AND NU_NOTA_MT IS NOT NULL
  AND NU_NOTA_REDACAO IS NOT NULL
GROUP BY TP_ESCOLA
ORDER BY Media_Geral DESC;

"""
pd.set_option("display.max_rows", None)

df_resultado = pd.read_sql_query(query, conn)
df_resultado
```
